```python
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Crear series de tiempo
ts_gasolina_super = pd.Series(df_consumo['Gasolina superior'].values, index=pd.to_datetime(df_consumo['Fecha']))

ts_gasolina_regular = pd.Series(df_consumo['Gasolina regular'].values, index=pd.to_datetime(df_consumo['Fecha']))
ts_gasolina_super = ts_gasolina_super.dropna()
# Agrega otras series según sea necesario

# Gráficos de series de tiempo
ts_gasolina_super.plot(title='Serie de tiempo de Gasolina Super')
plt.show()

# Descomposición de la serie de tiempo
decomposition = sm.tsa.seasonal_decompose(ts_gasolina_super, model='additive')
decomposition.plot()
plt.show()

# Estacionariedad - Test de Dickey-Fuller
from statsmodels.tsa.stattools import adfuller

result = adfuller(ts_gasolina_super.dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Suponiendo que ts_gasolina_super ya está definido y limpio
# Verificar la estacionariedad
result = adfuller(ts_gasolina_super)
print('ADF Statistic:', result[0])
print('p-value:', result[1])

# Identificar parámetros d
# Si la serie no es estacionaria, diferenciar
d = 1  # Por ejemplo, si la serie no es estacionaria, se usa d=1

# Graficar ACF y PACF para determinar p y q
plot_acf(ts_gasolina_super)
plot_pacf(ts_gasolina_super)
plt.show()

# Basado en los gráficos, elige p y q
p = 19  # Ejemplo, debe ser determinado por el gráfico PACF
q = 3 # Ejemplo, debe ser determinado por el gráfico ACF

# Modelado con ARIMA
model_arima = ARIMA(ts_gasolina_super, order=(p, d, q))
model_arima_fit = model_arima.fit()
print(model_arima_fit.summary())
```

```python
import numpy as np
np.float_ = np.float64
from prophet import Prophet

# Preparar los datos para Prophet
df_prophet = df_consumo[['Fecha', 'Gasolina superior']].rename(columns={'Fecha': 'ds', 'Gasolina superior': 'y'})

# Ajustar el modelo
model_prophet = Prophet()
model_prophet.fit(df_prophet)

# Predicciones
future = model_prophet.make_future_dataframe(periods=36, freq='M')
forecast = model_prophet.predict(future)

# Graficar las predicciones
model_prophet.plot(forecast)
plt.show()
```

```python
# Predicciones con ARIMA para Gasolina Superior
predictions = model_arima_fit.get_forecast(steps=36)
predicted_values = predictions.predicted_mean

plt.figure(figsize=(12, 6))
plt.plot(ts_gasolina_super, label='Serie Histórica')
plt.plot(predicted_values, label='Predicción ARIMA')
plt.legend()
plt.show()
```

```python
# Preparar los datos para Prophet
df_prophet_precios = df_precios_melted[['Fecha', 'Precio']].rename(columns={'Fecha': 'ds', 'Precio': 'y'})

# Ajustar el modelo Prophet
model_prophet_precios = Prophet()
model_prophet_precios.fit(df_prophet_precios)

# Predicciones con Prophet
future_precios = model_prophet_precios.make_future_dataframe(periods=36, freq='M')
forecast_precios = model_prophet_precios.predict(future_precios)

# Graficar las predicciones de Prophet
model_prophet_precios.plot(forecast_precios)
plt.show()
```
